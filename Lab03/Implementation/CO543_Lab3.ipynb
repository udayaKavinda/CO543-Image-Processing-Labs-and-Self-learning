##1. Apply Mean filtering with mask size 3x3 and 5x5

```python
def mean(img,kernal_size):
  '''
    This funcation applies the mean filter to any given image and returns the 
    resulting image.

    Note : Kernal size should be odd.
  '''
  
  dividing_value = kernal_size**2

  col_start = int(kernal_size/2)
  col_end = len(img[0])-int(kernal_size/2)
  row_start = int(kernal_size/2)
  row_end = len(img[:,0]) - int(kernal_size/2)
  
  output = []
  for i in range(row_start,row_end+1):
    row = []
    for j in range(col_start,col_end+1):
      kernal = img[i-int(kernal_size/2):i+int(kernal_size/2)+1,j-int(kernal_size/2):j+int(kernal_size/2)+1]
      value = int(np.sum(kernal)/dividing_value)
      row.append(value)

    output.append(row)
    
    return np.array(output)
```

```python
cv2_imshow(img)
mean_filtered_img = mean(img,3)
cv2_imshow(mean_filtered_img)
```
